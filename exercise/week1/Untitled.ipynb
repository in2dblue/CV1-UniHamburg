```python
def diction(string):
    new_dict = {}
    for i in range(len(string)):
        if string[i] in new_dict:
            new_dict[string[i]] +=1
        else:
            new_dict[string[i]]=1
    print(new_dict)
```

```python
def rgb_to_gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2125 * r + 0.7154 * g + 0.0721 * b

    return gray
```

```python
gray = rgb_to_gray(im)
```
