```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
import imageio
from scipy.signal import convolve2d
from skimage.filters import gaussian 
```

```python
im = imageio.imread("../week3/woman.png")
```

```python
g1 = gaussian(im,sigma=2)
g2 = gaussian(im,sigma=5)
```

# Difference of Gaussian

```python
f = plt.figure(figsize=(15,10))
ax = f.add_subplot(131)
ax2 = f.add_subplot(132)
ax3 = f.add_subplot(133)

ax.imshow(g1,cmap="gray")
ax2.imshow(g2,cmap="gray")
ax3.imshow(g1-g2,cmap="gray")

plt.show()
```

# Differnce of box filter  

```python
dob1 = np.ones([43,43]) * (1/43**2)
```

```python
dob2 = np.ones([19,19]) * (1/19**2)
```

```python
pass1 = convolve2d(im,dob2,mode='same')
```

```python
# plt.imshow(pass1, cmap='gray')
# plt.show()
```

```python
pass2 = convolve2d(im, dob1, mode='same')
```

```python
# plt.imshow(pass2, cmap='gray')
# plt.show()
```

```python
f = plt.figure(figsize=(15,10))
ax = f.add_subplot(141)
ax2 = f.add_subplot(142)
ax3 = f.add_subplot(143)
ax4 = f.add_subplot(144)

ax.imshow(pass1,cmap="gray")
ax2.imshow(pass2,cmap="gray")
ax3.imshow(pass2-pass1,cmap="gray")
ax4.imshow((pass2-pass1).clip(0,255), cmap='gray')
plt.show()
```
