```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
# from scipy.signal import convolve2d, correlate2d
from skimage.color import rgb2gray
from skimage.feature import match_template
```

```python
im = imageio.imread("coco.jpg")
im_clock = imageio.imread("coco_clock.jpg")
```

```python
im_gray = rgb2gray(im)
im_clock_gray= rgb2gray(im_clock)
```

```python
im_gray.dtype, im_clock_gray.dtype
```

```python
f = plt.figure(figsize=(10,8))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)

ax.imshow(im_gray,cmap="gray")
ax2.imshow(im_clock_gray,cmap="gray")

plt.show()
```

```python
x = match_template(im_gray,im_clock_gray)
```

```python
plt.imshow(x, cmap='gray')
```

```python
np.unravel_index(x.argmax(), x.shape)
```

```python
y = match_template(im_gray, np.fliplr(im_clock_gray))
plt.imshow(y>0.15, cmap='gray')
```

```python
np.unravel_index(y.argmax(), y.shape)
```
